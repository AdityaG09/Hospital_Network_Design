```python
from flask import Flask,render_template,request
import csv

def read():
    with open('/Users/abhishekpaul/Documents/Networks/Patient.csv', 'r') as file:
        reader = csv.reader(file, delimiter = '\t')
        data=[]
        for row in reader:
            if(len(row)>0):
                data.append(row[0].split(','))
        return data
        
app = Flask(__name__)

ID = 'SRM Hospital'
password = 'binod'
clnts = {'Client 1':'abcd','Client 2':'1234','Client 3':'pass'}

@app.route('/',methods=['POST','GET'])
def server_home():
    if(request.form.get('id')!=None or request.form.get('pass')!=None):
        if(request.form.get('id')==ID and request.form.get('pass')==password):
            dat1=read()
            return render_template('server_home.html',dat=dat1)
        else:
            return render_template('server_error.html')
    else:
        return render_template('server_login.html')
    
@app.route('/client/',methods=['POST','GET'])
def client_home():
    if(request.form.get('id')!=None or request.form.get('pass')!=None):
        if(request.form.get('id') in clnts.keys() and request.form.get('pass')==clnts[request.form.get('id')]):
            return render_template('client_home.html',cid=request.form.get('id'))
        else:
            return render_template('client_error.html')
    else:
        return render_template('client_login.html')
        
@app.route('/add/',methods=['POST'])
def add():
    name = request.form.get('nam')
    age = request.form.get('age')
    con = request.form.get('con')
    add = request.form.get('add')
    med = request.form.get('med')
    idstr = genID(con)
    newRow = '\n'+idstr+','+name+','+age+','+con+','+add+','+med
    with open('/Users/abhishekpaul/Documents/Networks/Patient.csv','a') as fd:
        fd.write(newRow)
    dat1 = read()
    return render_template('server_home.html',dat=dat1)

@app.route('/delete/<id>',methods=['GET'])
def delete(id):
    data = read()
    for i in data:
        if(i[0]==id): 
            data.remove(i)
    with open('/Users/abhishekpaul/Documents/Networks/Patient.csv', 'w') as writeFile:
        writer = csv.writer(writeFile)
        writer.writerows(data)
    return render_template('server_home.html',dat=data)

@app.route('/pid/<cl>',methods=['GET','POST'])
def pid(cl):
    key = request.form.get('id')
    data = read()
    ans=[]
    for i in data:
        if(i[0]==key):
            ans.append(i)
            break
    return render_template('client_data.html',dat=ans,cid=cl)

@app.route('/age/<cl>',methods=['GET','POST'])
def age(cl):
    rang = request.form.get('range')
    key = request.form.get('age')
    data = read()
    ans=[]
    for i in data:
        if(rang=='equal' and i[2]==key):
            ans.append(i)
            continue
        elif(rang=='less' and i[2]<key):
            ans.append(i)
            continue
        elif(rang=='more' and i[2]>key):
            ans.append(i)
    return render_template('client_data.html',dat=ans,cid=cl)


@app.route('/med/<cl>',methods=['GET','POST'])
def med(cl):
    key = request.form.get('med')
    data = read()
    ans=[]
    for i in data:
        if(i[5]==key):
            ans.append(i)
    return render_template('client_data.html',dat=ans,cid=cl)

def genID(con):
    data=read()
    ids=[]
    for i in data:
        ids.append(i[0])
    temp = int(con[7:10])
    while('P'+str(temp) in ids):
        temp+=1
    return ('P'+str(temp))

app.run('0.0.0.0', 10001)
```
